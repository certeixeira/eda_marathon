**Como deu poucos dados, vou pegar de outro país, no caso USA**

```python
df_usa = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'])]
```

```python
# como meus 16gb de ram não são o suficiente para rodar o código acima, vou ter que filtrar em duas partes
# Aplicar os primeiros filtros
df_temp = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]

# Aplicar o último filtro
df_usa = df_temp[df_temp['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
df_usa.head(10)
```

```python
df_usa.shape
```

Como sabemos que os dados são dos USA, vamos manter apenas o nome dos eventos

```python
# Remove (USA) do nome dos eventos
```

```python
df_usa['Event name'] = df_usa['Event name'].str.split('(').str.get(0)
df_usa.head(10)
```

**Como a coluna 'Athlete age category' não é precisa, vamos adicionar uma coluna chama 'Athlete age' que mostra a idade do atleta em 2020**

```python
df_usa['Athlete age'] = 2020 - df_usa['Athlete year of birth']
df_usa
```

**Vamos remover o 'h' de hora da coluna 'Athlete performance'**

```python
df_usa['Athlete performance'] = df_usa['Athlete performance'].str.split(' ').str.get(0)
```

**Removendo colunas desnecessárias e valores nulos**

```python
# apaga colunas Athlete club, Athlete country, Athlete year of birth, Athlete age category
df_usa = df_usa.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
df_usa
```

```python
df_usa[df_usa['Athlete age'].isna()==1]
```

```python
# mudando o 'Athlete age' para int
df_usa['Athlete age'] = df_usa['Athlete age'].astype(int)
```

## Qual a diferença na velocidade dos atletas masculinos e femininos em corridas de 50km e 50mi?

```python
df_usa.groupby(['distancia_corrida', 'genero_atleta'])['velocidade_media_atleta'].mean()
```

## Quais grupos de idade se sobressaem nas corridas de 50mi?

```python
# o count tem que ser >= a 20
df_usa.query('distancia_corrida == "50mi"').groupby('idade_atleta')['velocidade_media_atleta'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19')
```

## Quais grupos de idade se sobressaem nas corridas de 50km?

```python
# o count tem que ser >= a 20
df_usa.query('distancia_corrida == "50km"').groupby('idade_atleta')['velocidade_media_atleta'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19')
```

## A velocidade média no verão é mais rápida que no inverno?

**Precisamos criar duas novas colunas, a primeira mostra o mês em que a corrida aconteceu**

```python
df_usa['mes_corrida'] = df_usa['data_corrida'].str.split('.').str.get(1).astype(int)
df_usa
```

**Em seguida criamos uma coluna que mostra a estação em que ocorreu a corrida de acordo com  o mês**

```python
df_usa['estacao_corrida'] = df_usa['mes_corrida'].apply(lambda x: 'inverno' if x > 11 else 'outono' if x > 8 else 'verão' if x > 5 else 'primavera' if x > 2 else 'inverno')
df_usa
```

**Chegamos ao seguinte resultado:**

```python
df_usa.groupby('estacao_corrida')['velocidade_media_atleta'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

**Separando o resultado para corridas de 50km**

```python
df_usa.query('distancia_corrida == "50km"').groupby('estacao_corrida')['velocidade_media_atleta'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
